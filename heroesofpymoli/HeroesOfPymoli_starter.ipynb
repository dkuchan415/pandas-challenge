```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "../heroesofpymoli/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Create a list of unique items

unique_items = purchase_data["Item Name"].unique()
unique_items

#Calculate average price

average_item_price = purchase_data["Price"].mean()
average_item_price

#Calculate number of purchases

purchases = purchase_data["SN"].count()
purchases

#Calculate total revenue

total_revenue = purchase_data["Price"].sum()
total_revenue


#Create table with new metrics

summary = pd.DataFrame({
    "Number of Unique Items":[len(unique_items)], 
    "Average Price":[average_item_price],
    "Number of Purchases":[purchases],
    "Total Purchases":[total_revenue]
})

#Format cells as currency

summary["Average Price"] = summary["Average Price"].map("${:,.2f}".format)
summary["Total Purchases"] = summary["Total Purchases"].map("${:,.2f}".format)

summary.head()
```
